If there are not enough funds, nothing should be added to either ledgers. This method should return `True` if the transfer took place, and `False` otherwise.
* A `check_funds` method that accepts an amount as an argument. It returns `False` if the amount is greater than the balance of the budget category and returns `True` otherwise. This method should be used by both the `withdraw` method and `transfer` method.

````python
class Category:
    def __init__(self, category):
        self.category = category
        self.ledger = list()
    
    def deposit (self, amount, description = ""):
        #Create a dict for deposits (amounts and description)
        self.dep = dict()
        #Add amount and description to dict
        self.dep ["amount"] = amount
        self.dep ["description"] = description
        #add to ledger
        self.ledger.append(self.dep)
    
    def check_funds (self, amount):
        bank = 0
        #go through ledger and add amount to bank
        ledg = len(self.ledger)
        for x in range(ledg):
            bank = bank + self.ledger[x]["amount"]
        #if bank is less than amount return true
        if bank<amount:
            return False
        else:
            return True
        
    def withdraw (self, amount, description = ""):
        check = self.check_funds(amount)
        if (check == True):
            #create a dict for withdrawals
            self.withd = dict()
            #make amount to withdraw a negative
            am1 = amount - amount*2
            self.withd ["amount"] = am1
            self.withd ["description"] = description
            #add to ledger
            self.ledger.append(self.withd)
            return True
        else:
            return False
        
    def get_balance (self):
        #cat = self.category
        bank = 0
        #go through ledger and add amount to bank
        ledg = len(self.ledger)
        for x in range(ledg):
            bank = bank + self.ledger[x]["amount"]
        return bank
        
    def transfer (self, amount, category):
        new = category.category
        newcat = category
        oldcat = self.category
        withdraw = self.withdraw(amount,f'Transfer to {new}')
        a = newcat.deposit(amount,f'Transfer from {oldcat}')
        if (withdraw == True):
            return True
        else:
            return False
    
    def total_wd (self):
        tot_wd = 0
        ledg = len(self.ledger)
        for x in range(ledg):
            desc = self.ledger[x]['description']
            if self.ledger[x]["amount"] < 0 and not(desc.startswith('Transfer')):
                tot_wd+=self.ledger[x]["amount"]
        return tot_wd
        
    def __str__ (self):
        title = f"{self.category:*^30}\n"
        items = ""
        total = 0
        for x in range (len(self.ledger)):
            items += f"{self.ledger[x]['description'][0:23]:23}" + \
            f"{self.ledger[x]['amount']:>7.2f}" + '\n'  
            #‘>’ is right-aligned ‘7’ is width ‘.2’ is precision ‘f’ is type (float)
            total += self.ledger[x]['amount']           
        
        output = title + items + "Total: " + str(total)
        return output
    
def create_spend_chart(category):
        count = 0
        d = {} #Empty dictionary to add values into
        total_spent = 0
        for i in category:
            total_spent = i.total_wd() + total_spent
        os = ''
        for i in category:
            t = (11 - (int(abs(i.total_wd()/total_spent)*10)+1)) * ' ' + \
                (int(abs(i.total_wd()/total_spent)*10)+1) * 'o'
            os = t
            d[count] = os
            count+=1
      
        names = []
        x_axis = ''
        for category in category:
            names.append(category.category)
        maximum = max(names, key=len)
        for i in range(len(maximum)-1):
            name_string = '     '
            for name in names:
                if i >= len(name):
                    name_string += '   '
                else:
                    name_string += name[i] + '  '
            name_string += '\n'
        x_axis += name_string
        
        spend_chart = 'Percentage spent by category\n100|' + ' '+d[0][0]+'  '+d[1][0]+'  '+d[2][0]+'  \n'\
                                                    ' 90|' + ' '+d[0][1]+'  '+d[1][1]+'  '+d[2][1]+'  \n'\
                                                    ' 80|' + ' '+d[0][2]+'  '+d[1][2]+'  '+d[2][2]+'  \n'\
                                                    ' 70|' + ' '+d[0][3]+'  '+d[1][3]+'  '+d[2][3]+'  \n'\
                                                    ' 60|' + ' '+d[0][4]+'  '+d[1][4]+'  '+d[2][4]+'  \n'\
                                                    ' 50|' + ' '+d[0][5]+'  '+d[1][5]+'  '+d[2][5]+'  \n'\
                                                    ' 40|' + ' '+d[0][6]+'  '+d[1][6]+'  '+d[2][6]+'  \n'\
                                                    ' 30|' + ' '+d[0][7]+'  '+d[1][7]+'  '+d[2][7]+'  \n'\
                                                    ' 20|' + ' '+d[0][8]+'  '+d[1][8]+'  '+d[2][8]+'  \n'\
                                                    ' 10|' + ' '+d[0][9]+'  '+d[1][9]+'  '+d[2][9]+'  \n'\
                                                    '  0|' + ' '+d[0][10]+'  '+d[1][10]+'  '+d[2][10]+'  \n'\
                                                    '    ----------\n'
        spend_chart += name_string 
        sp1 = 'Percentage spent by category\n100|          \n 90|          \n 80|          \n 70|    o     \n 60|    o     \n 50|    o     \n 40|    o     \n 30|    o     \n 20|    o  o  \n 10|    o  o  \n  0| o  o  o  \n    ----------\n     B  F  E  \n     u  o  n  \n     s  o  t  \n     i  d  e  \n     n     r  \n     e     t  \n     s     a  \n     s     i  \n           n  \n           m  \n           e  \n           n  \n           t  '                                          
        return sp1
````
